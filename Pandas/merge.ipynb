```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
df3=pd.merge(df1,df2,on="city")
df3
```

```python
df4 = pd.DataFrame({
    "city": ["new york","chicago","orlando","baltimore"],
    "temperature": [21,14,35,45],
})
df4
```

```python
df5 = pd.DataFrame({
    "city": ["chicago","new york","san fransisco"],
    "humidity": [65,68,75],
})
df5
```

```python
df6=pd.merge(df4,df5,on="city")
df6
```

```python
df6=pd.merge(df4,df5,on="city",how="outer",indicator=True)
df6
```

```python
df6=pd.merge(df4,df5,on="city",how="inner")
df6
```

```python
df6=pd.merge(df4,df5,on="city",how="left")
df6
```

```python
df6=pd.merge(df4,df5,on="city",how="right")
df6
```

suffixes

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df2
```

```python
df3= pd.merge(df1,df2,on="city",how="outer", suffixes=('_first','_second'))
df3
```

Join

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1.set_index('city',inplace=True)
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2.set_index('city',inplace=True)
df2
```

```python
df1.join(df2,lsuffix='_l', rsuffix='_r')
```
